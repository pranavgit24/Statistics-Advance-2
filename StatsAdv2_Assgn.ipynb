```python
#Q3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution

def bernoulli_pmf(x, p):
    """
    Calculates the PMF of a Bernoulli distribution.

    Parameters:
        x (int): The outcome (0 or 1)
        p (float): Probability of success (0 ≤ p ≤ 1)

    Returns:
        float: Probability of observing outcome x
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("Probability p must be between 0 and 1.")

    return p**x * (1 - p)**(1 - x)


print(bernoulli_pmf(1, 0.7))  # Output: 0.7
print(bernoulli_pmf(0, 0.7))  # Output: 0.3
```

```python
#Q9 Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores for a dataset and plots them.

    Parameters:
        data (list or array): Input numeric data
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot the Z-scores
    plt.figure(figsize=(8, 5))
    plt.plot(z_scores, 'o', color='teal', label='Z-scores')
    plt.axhline(y=0, color='gray', linestyle='--')
    plt.axhline(y=1, color='red', linestyle=':', label='±1 SD')
    plt.axhline(y=-1, color='red', linestyle=':')
    plt.title("Z-scores of Dataset")
    plt.xlabel("Data Point Index")
    plt.ylabel("Z-score")
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.legend()
    plt.show()

    return z_scores

data = [65, 70, 75, 80, 85, 90, 95]
z_scores = calculate_and_plot_z_scores(data)
print("Z-scores:", z_scores)
```

```python
#Q14 Write a Python program to calculate the z-score for a given data point and compare it to a standard normal distribution

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std_dev):
    """
    Calculate the Z-score for a data point.

    Parameters:
        x (float): The data point
        mean (float): Mean of the distribution
        std_dev (float): Standard deviation of the distribution

    Returns:
        float: Z-score
    """
    z = (x - mean) / std_dev
    return z

def plot_standard_normal_with_z(z):
    """
    Plot the standard normal distribution and mark the given Z-score.

    Parameters:
        z (float): Z-score to highlight
    """
    x = np.linspace(-4, 4, 500)
    pdf = norm.pdf(x)

    plt.figure(figsize=(10, 5))
    plt.plot(x, pdf, label='Standard Normal Distribution', color='navy')
    plt.fill_between(x, 0, pdf, where=(x <= z), color='skyblue', alpha=0.6, label=f'P(Z ≤ {z:.2f})')
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    plt.title("Standard Normal Distribution with Z-score")
    plt.xlabel("Z-score")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()

# Example usage
x = 85         # data point
mean = 75      # sample/population mean
std_dev = 5    # standard deviation

z = calculate_z_score(x, mean, std_dev)
print(f"Z-score for x = {x}: {z:.2f}")

plot_standard_normal_with_z(z)
```

```python
#Q15 Implement hypothesis testing using Z-statistics for a sample dataset

import numpy as np
from scipy.stats import norm

# Step 1: Input the data
sample_mean = 980           # sample mean (x̄)
population_mean = 1000      # claimed population mean (μ₀)
sample_std = 50             # sample standard deviation (s)
n = 40                      # sample size
alpha = 0.05                # significance level

# Step 2: Calculate the Z-statistic
standard_error = sample_std / np.sqrt(n)
z_stat = (sample_mean - population_mean) / standard_error

# Step 3: Calculate the p-value (one-tailed, left-side)
p_value = norm.cdf(z_stat)

# Step 4: Make decision
print(f"Z-statistic: {z_stat:.3f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: Evidence suggests the bulbs last less than 1000 hours.")
else:
    print("Fail to reject the null hypothesis: No strong evidence bulbs last less than 1000 hours.")
```

```python
#Q16 Create a confidence interval for a dataset using Python and interpret the result

import numpy as np
from scipy.stats import norm

# Step 1: Sample dataset (e.g., product weights in grams)
data = [50.2, 49.7, 51.0, 50.5, 49.9, 50.1, 50.4, 50.3, 49.8, 50.0]

# Step 2: Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)   # use ddof=1 for sample standard deviation
n = len(data)

# Step 3: Set confidence level (e.g., 95%)
confidence_level = 0.95
alpha = 1 - confidence_level

# Step 4: Calculate Z critical value for 95% CI
z_critical = norm.ppf(1 - alpha/2)

# Step 5: Calculate standard error and margin of error
standard_error = sample_std / np.sqrt(n)
margin_of_error = z_critical * standard_error

# Step 6: Compute confidence interval
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

# Display results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval: ({lower_bound:.2f}, {upper_bound:.2f})")
```

```python
#Q24 Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution

import numpy as np
from scipy.stats import skew, kurtosis

def analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000):
    """
    Generates data from a normal distribution and calculates skewness and kurtosis.

    Parameters:
        mean (float): Mean of the distribution
        std (float): Standard deviation
        size (int): Number of samples

    Returns:
        dict: Skewness and kurtosis of the dataset
    """
    # Generate data
    data = np.random.normal(loc=mean, scale=std, size=size)

    # Calculate skewness and kurtosis
    skew_val = skew(data)
    kurt_val = kurtosis(data, fisher=False)  # use fisher=False to get Pearson's definition (normal → 3)

    # Print results
    print(f"Skewness: {skew_val:.4f}")
    print(f"Kurtosis: {kurt_val:.4f}")

    return {"skewness": skew_val, "kurtosis": kurt_val}
analyze_normal_distribution_skew_kurtosis(mean=0, std=1, size=1000)
```

```python
#Q1 Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results

import numpy as np
from scipy import stats

# Sample data (e.g., test scores from a sample group)
sample_data = [72, 75, 78, 74, 73, 77, 76, 74, 75, 79]

# Known population parameters
population_mean = 70
population_std = 5  # Known population standard deviation

# Sample statistics
sample_mean = np.mean(sample_data)
sample_size = len(sample_data)

# Z-test calculation
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))
p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))  # Two-tailed test

# Output
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-score: {z_score:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The sample mean is significantly different from the population mean.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference from the population mean.")
```

```python
#Q4 Perform a two-tailed Z-test using Python and visualize the decision region on a plot

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Step 1: Simulate sample data
np.random.seed(0)
sample_data = np.random.normal(loc=102, scale=10, size=50)

# Step 2: Population parameters
population_mean = 100
population_std = 10  # Known population standard deviation
n = len(sample_data)

# Step 3: Calculate sample mean and Z-score
sample_mean = np.mean(sample_data)
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))

# Step 4: Calculate two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z_score)))

# Step 5: Print result
print(f"Sample Mean   : {sample_mean:.2f}")
print(f"Z-score       : {z_score:.4f}")
print(f"P-value       : {p_value:.4f}")

alpha = 0.05
z_critical = norm.ppf(1 - alpha/2)  # two-tailed critical z-value

if abs(z_score) > z_critical:
    print("Result: Reject H₀ (Significant difference)")
else:
    print("Result: Fail to reject H₀ (No significant difference)")

# Step 6: Plot the Z-distribution and decision regions
z_vals = np.linspace(-4, 4, 1000)
pdf_vals = norm.pdf(z_vals)

plt.figure(figsize=(10, 6))
plt.plot(z_vals, pdf_vals, label='Standard Normal Distribution')

# Fill critical regions
plt.fill_between(z_vals, 0, pdf_vals, where=(z_vals <= -z_critical), color='red', alpha=0.4, label='Rejection Region (Left)')
plt.fill_between(z_vals, 0, pdf_vals, where=(z_vals >= z_critical), color='red', alpha=0.4, label='Rejection Region (Right)')

# Plot observed z-score
plt.axvline(z_score, color='blue', linestyle='--', linewidth=2, label=f'Z-score = {z_score:.2f}')

plt.title("Two-tailed Z-test Decision Regions")
plt.xlabel("Z-value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Q5 Create a Python function that calculates and visualizes Type 1 and Type 2 errors during Hypothesis testing

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05):
    """
    Visualize Type I and Type II errors for a right-tailed hypothesis test.
    Parameters:
        mu0 : mean under null hypothesis (H₀)
        mu1 : mean under alternative hypothesis (H₁)
        sigma : population standard deviation
        n : sample size
        alpha : significance level
    """
    # Standard error
    se = sigma / np.sqrt(n)

    # Critical value (right-tailed test)
    z_critical = norm.ppf(1 - alpha)
    x_critical = mu0 + z_critical * se

    # Beta (Type II error): probability of not rejecting H₀ when H₁ is true
    z_beta = (x_critical - mu1) / se
    beta = norm.cdf(z_beta)

    # Visualization range
    x = np.linspace(mu0 - 4*se, mu1 + 4*se, 1000)

    # Distributions
    y0 = norm.pdf(x, mu0, se)  # Under H₀
    y1 = norm.pdf(x, mu1, se)  # Under H₁

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y0, label='H₀ Distribution (μ₀)', color='blue')
    plt.plot(x, y1, label='H₁ Distribution (μ₁)', color='green')

    # Type I Error region (α): right tail of H₀
    plt.fill_between(x, 0, y0, where=(x >= x_critical), color='red', alpha=0.3, label='Type I Error (α)')

    # Type II Error region (β): left of x_critical under H₁
    plt.fill_between(x, 0, y1, where=(x <= x_critical), color='orange', alpha=0.3, label='Type II Error (β)')

    # Critical value line
    plt.axvline(x_critical, color='black', linestyle='--', label=f'Critical Value = {x_critical:.2f}')

    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Sample Mean')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Print numeric results
    print(f"Critical Value (xₐ): {x_critical:.2f}")
    print(f"Type I Error (α): {alpha:.4f}")
    print(f"Type II Error (β): {beta:.4f}")
    print(f"Power of the test (1 - β): {1 - beta:.4f}")

# Example usage
visualize_type1_type2_errors(mu0=100, mu1=105, sigma=10, n=30, alpha=0.05)
```

```python
#Q6 Write a Python program to perform an independent T-test and interpret the results

import numpy as np
from scipy import stats

# Simulated data: scores of two independent groups
group1 = np.array([88, 92, 85, 90, 87, 91, 89])
group2 = np.array([82, 84, 79, 85, 83, 80, 81])

# Perform independent two-sample t-test
t_stat, p_value = stats.ttest_ind(group1, group2, equal_var=False)  # Welch’s t-test

# Output results
print("Group 1 Mean:", np.mean(group1))
print("Group 2 Mean:", np.mean(group2))
print(f"T-statistic : {t_stat:.4f}")
print(f"P-value     : {p_value:.4f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis. The group means are significantly different.")
else:
    print("Result: Fail to reject the null hypothesis. No significant difference in means.")
```

```python
#Q9 Write a Python function to calculate the confidence interval for a sample mean and explain its significance

import numpy as np
from scipy import stats

def confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
        data       : array-like, the sample data
        confidence : confidence level (default is 0.95 for 95%)

    Returns:
        (lower_bound, upper_bound)
    """
    n = len(data)
    mean = np.mean(data)
    std_err = stats.sem(data)  # standard error of the mean
    margin = stats.t.ppf((1 + confidence) / 2, df=n-1) * std_err
    return mean - margin, mean + margin

"""
What the Confidence Interval Means:
A 95% confidence interval means that if you took many random samples and computed an interval from each,
about 95% of those intervals would contain the true population mean.

It is not a probability that the true mean lies within the interval — the true mean is fixed,
but the interval is estimated from sample data.
"""


# Example usage:
np.random.seed(1)
sample = np.random.normal(loc=100, scale=15, size=30)
ci_low, ci_high = confidence_interval(sample)

print(f"95% Confidence Interval for the Mean: ({ci_low:.2f}, {ci_high:.2f})")

```

```python
#Q10 Write a Python program to calculate the margin of error for a given confidence level using sample data

import numpy as np
from scipy import stats

def calculate_margin_of_error(data, confidence=0.95):
    """
    Calculate the margin of error for a given confidence level.

    Parameters:
        data       : array-like, the sample data
        confidence : desired confidence level (default is 95%)

    Returns:
        margin of error (float)
    """
    n = len(data)
    std_err = stats.sem(data)  # standard error = s / sqrt(n)
    t_score = stats.t.ppf((1 + confidence) / 2, df=n-1)  # t critical value
    margin = t_score * std_err
    return margin

# Example usage
np.random.seed(42)
sample_data = np.random.normal(loc=50, scale=8, size=40)

margin = calculate_margin_of_error(sample_data, confidence=0.95)
print(f"Margin of Error (95% confidence): ±{margin:.2f}")
```

```python
#Q11 Implement a Bayesian inference method using Bayes' theorem in Python and explain the process

def bayesian_inference(prior, likelihood, false_positive_rate):
    """
    Calculates posterior using Bayes' theorem.

    Parameters:
        prior (P(H))                 : Prior probability (e.g., disease prevalence)
        likelihood (P(Pos|H))        : Probability of positive test given disease
        false_positive_rate (P(Pos|¬H)): Probability of positive test without disease

    Returns:
        Posterior probability: P(H|Pos)
    """
    # Complement of prior: P(not H)
    prior_complement = 1 - prior

    # Total probability of testing positive (P(D))
    marginal = (likelihood * prior) + (false_positive_rate * prior_complement)

    # Bayes' Theorem: Posterior
    posterior = (likelihood * prior) / marginal
    return posterior

# Example parameters
prior = 0.01                # P(Disease)
likelihood = 0.95           # P(Pos | Disease)
false_positive = 0.05       # P(Pos | No Disease)

posterior = bayesian_inference(prior, likelihood, false_positive)

print(f"Probability of actually having the disease given a positive test: {posterior:.4f}")
```

```python
#Q12 Perform a Chi-square test for independence between two categorical variables in Python

import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Create a contingency table
# Columns: Product Preference (Yes / No)
# Rows: Gender (Male / Female)
data = np.array([[30, 10],   # Male: 30 said Yes, 10 said No
                 [20, 20]])  # Female: 20 said Yes, 20 said No

# Convert to DataFrame for readability
df = pd.DataFrame(data,
                  index=['Male', 'Female'],
                  columns=['Yes', 'No'])

# Perform Chi-square test
chi2, p, dof, expected = chi2_contingency(df)

# Display Results
print("Contingency Table:")
print(df)
print("\nExpected Frequencies (if independent):")
print(pd.DataFrame(expected, index=df.index, columns=df.columns))
print(f"\nChi-square Statistic : {chi2:.4f}")
print(f"Degrees of Freedom    : {dof}")
print(f"P-value               : {p:.4f}")

# Interpret the result
alpha = 0.05
if p < alpha:
    print("Result: Reject the null hypothesis — variables are dependent.")
else:
    print("Result: Fail to reject the null hypothesis — variables are independent.")
```

```python
#Q13 Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data

import numpy as np
import pandas as pd

# Step 1: Observed data as a contingency table
observed = np.array([
    [50, 30],
    [20, 40]
])

# Step 2: Convert to DataFrame for clarity
observed_df = pd.DataFrame(observed,
                           index=['Group A', 'Group B'],
                           columns=['Outcome X', 'Outcome Y'])

# Step 3: Calculate expected frequencies manually
row_totals = observed_df.sum(axis=1).values.reshape(-1, 1)  # shape (rows, 1)
col_totals = observed_df.sum(axis=0).values.reshape(1, -1)  # shape (1, cols)
grand_total = observed_df.values.sum()

expected = (row_totals @ col_totals) / grand_total  # matrix multiplication

# Step 4: Display results
expected_df = pd.DataFrame(expected, index=observed_df.index, columns=observed_df.columns)

print("Observed Frequencies:")
print(observed_df)
print("\nExpected Frequencies (assuming independence):")
print(expected_df.round(2))
```

# THEORY QUESTIONS (Statistics Part 2)

# Q 1: What is hypothesis testing in statistics?

ANS: Hypothesis testing in statistics is a formal method used to make decisions or inferences about population parameters based on sample data.

Hypothesis testing is a statistical process for testing an assumption (called a hypothesis) about a population parameter using sample data.

#Key Concepts:

1. Null Hypothesis (H₀):

> The default assumption or status quo.

> Example: "The average height of adult women is 160 cm."

2. Alternative Hypothesis (H₁ or Ha):

> The claim you want to test or prove.

> Example: "The average height of adult women is not 160 cm."

3. Test Statistic:

> A value calculated from the sample data that is used to test the hypothesis (e.g., Z-score, t-score).

4. Significance Level (α):

> The probability of rejecting the null hypothesis when it is true (Type I error).

> Common values: 0.05, 0.01.

5. P-value:

> The probability of observing the sample result, or more extreme, if H₀ is true.

> If p-value < α, reject H₀.

6. Decision:

> Based on the p-value and α, we either:

>> Reject H₀ → evidence supports H₁

>> Fail to reject H₀ → not enough evidence to support H₁

# Purpose of Hypothesis Testing:

> To make data-driven decisions

> To test theories or claims

> To evaluate the effectiveness of treatments/interventions

# Example:

Suppose a company claims that their batteries last 20 hours on average.

You suspect it's less.

> H₀: μ = 20 (mean battery life is 20 hours)

> H₁: μ < 20 (mean battery life is less than 20 hours)

> You collect a sample of 30 batteries and compute the mean and standard deviation.

> Perform a one-sample t-test.

> If p-value < 0.05 → reject H₀ → support your suspicion.

# Q 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

ANS:
# Null Hypothesis (H₀):

> The null hypothesis is the default or original assumption we make about a population.

> It usually states that there is no effect, no difference, or no relationship.

> It’s what you test against.

Purpose: Acts as a starting point for statistical testing.

Assumes: The observed data occurred by chance.

# Example:
H₀: The new drug has no effect on blood pressure.

(Mathematically: H₀: μ = μ₀)

# Alternative Hypothesis (H₁ or Ha):

> The alternative hypothesis is what you want to prove.

> It suggests there is an effect, a difference, or a relationship.

> It directly contradicts H₀.

Purpose: Represents the outcome if H₀ is false.

Assumes: The observed data reflects a real effect, not just chance.

# Example:

H₁: The new drug lowers blood pressure.

(Mathematically: H₁: μ < μ₀)

# Example in Context:

Suppose a manufacturer claims their bulbs last 1000 hours.
> H₀: μ = 1000 (bulbs last 1000 hours on average)

> H₁: μ ≠ 1000 (bulbs do not last 1000 hours)

You collect a sample and test the claim. If the data strongly disagrees with H₀ (p-value < α), you reject H₀ in favor of H₁.

| Aspect            | Null Hypothesis (H₀)            | Alternative Hypothesis (H₁)  |
| ----------------- | ------------------------------- | ---------------------------- |
| Assumes           | No change or effect             | There is a change or effect  |
| Goal              | To test whether to reject it    | To prove it (indirectly)     |
| Decision          | Reject or fail to reject        | Never "accept", only support |
| Type I Error (α)  | Rejecting H₀ when it is true    | False positive               |
| Type II Error (β) | Failing to reject H₀ when false | False negative               |

# Q 3: What is the significance level in hypothesis testing, and why is it important?

ANS: The significance level, denoted by α (alpha), is the threshold probability used to decide whether to reject the null hypothesis (H₀).
The significance level (α) is the maximum probability of making a Type I error, i.e., rejecting a true null hypothesis.

# Why Is It Important?

1. Controls Error Rate:

> α sets a limit on how often we make false discoveries (Type I errors).

2. Decision Rule Basis:

> If the p-value ≤ α, we reject H₀.

> If the p-value > α, we fail to reject H₀.

3. Determines Test Rigor:

> Lower α → more cautious (less likely to reject H₀).

> Higher α → more willing to detect effects (but higher false positive risk).

# Common Values of α:

| α Value | Meaning                             |
| ------- | ----------------------------------- |
| 0.05    | 5% chance of rejecting a true H₀    |
| 0.01    | 1% chance (stricter criterion)      |
| 0.10    | 10% chance (more lenient criterion) |

# Example:

Suppose you're testing if a new medicine is more effective than the standard one:

> H₀: No difference in effectiveness.

> H₁: New medicine is more effective.

> You set α = 0.05 (5% risk of falsely saying it's more effective).

> After calculations, p-value = 0.03

Since 0.03 < 0.05, you reject H₀ and conclude there's evidence the new medicine is more effective.

# Q 4: What does a P-value represent in hypothesis testing?

ANS: The p-value (probability value) is one of the most important concepts in hypothesis testing.
The p-value is the probability of obtaining test results at least as extreme as the observed results, assuming the null hypothesis (H₀) is true.

# Interpretation:

> It measures how compatible your data is with the null hypothesis.

> Smaller p-value → less likely the observed result occurred by random chance → stronger evidence against H₀.

# Decision Rule (based on significance level α):

| **If p-value ≤ α** | Reject H₀ → Evidence supports H₁        |
| ------------------ | --------------------------------------- |
| **If p-value > α** | Fail to reject H₀ → Not enough evidence |

Typical α values: 0.05, 0.01, 0.10

# How to Interpret P-values?

| P-value       | Interpretation                  |
| ------------- | ------------------------------- |
| **> 0.10**    | Weak evidence against H₀        |
| **0.05–0.10** | Moderate evidence against H₀    |
| **0.01–0.05** | Strong evidence against H₀      |
| **< 0.01**    | Very strong evidence against H₀ |

# Example:

You're testing whether a new teaching method improves student scores.

> H₀: The new method has no effect.

> H₁: The new method improves scores.

> After testing, you get a p-value = 0.02

> If α = 0.05 → 0.02 < 0.05, so you reject H₀

Conclusion: There’s significant evidence the new method helps.

# Q 5: How do you interpret the P-value in hypothesis testing?

ANS: The p-value helps you decide whether to reject or fail to reject the null hypothesis (H₀).

# Core Interpretation:
The p-value is the probability of getting your observed result (or one more extreme), assuming H₀ is true.

# Decision Rules (using significance level α):

| **Condition**   | **Interpretation**             | **Conclusion**        |
| --------------- | ------------------------------ | --------------------- |
| **p-value ≤ α** | Strong evidence **against H₀** | **Reject H₀**         |
| **p-value > α** | Not enough evidence against H₀ | **Fail to reject H₀** |

Common α values:

> 0.05 (5%) — most common

> 0.01 (1%) — more strict

> 0.10 (10%) — more lenient

# Interpretation Examples:
# Example 1:

> Hypothesis: A new drug lowers blood pressure.

> Result: p-value = 0.03

0.03 < 0.05 → Reject H₀

# Q 6: What are Type 1 and Type 2 errors in hypothesis testing?

ANS: In hypothesis testing, we make decisions based on sample data — but there's always a chance of making an error.

# Errors are classified into two types:

1. Type I Error (False Positive)

Rejecting the null hypothesis (H₀) when it is actually true.

> You're saying there's an effect or difference — but there really isn't.

> Think: "Detecting something that doesn't exist."

# Example:

You test a new drug and conclude it works (reject H₀), but in reality, it's no better than the placebo (H₀ was true).

> Probability of Type I Error = α (significance level)

> Common choices: α = 0.05 or 0.01

2. Type II Error (False Negative)

Failing to reject the null hypothesis when it is actually false.

> You're saying there’s no effect or difference — but there actually is.

> Think: "Missing something that does exist."

# Example:

You test a drug and conclude it doesn’t work (fail to reject H₀), but in reality, it does have an effect (H₀ was false).

> Probability of Type II Error = β

> Power of the test = 1 - β → Probability of correctly rejecting a false H₀

# Trade-off Between Type I and Type II Errors:

> Lowering α (to reduce Type I error) increases risk of Type II error, and vice versa.

> Researchers must balance both, depending on context:

>> In medical testing, Type I error can cause false alarm (e.g., saying someone has cancer when they don't),

>> Type II error can cause a missed diagnosis (e.g., failing to detect cancer when it's there).

# Example in Practice:

Testing whether a machine produces bolts with correct diameter (H₀: diameter = 5mm):

> Type I Error: You stop the machine, thinking it's faulty, but it was actually fine.

> Type II Error: You keep the machine running, even though it's actually producing wrong-sized bolts.

# Q 7: What is the difference between a one-tailed and a two-tailed test in hypothesis testing?

ANS: In hypothesis testing, the difference between a one-tailed and a two-tailed test lies in the direction of the hypothesis.

# How the rejection region is defined:

# One-Tailed Test:

> Tests for an effect in one specific direction.

> The alternative hypothesis (H₁) states that a parameter is either greater than or less than a certain value.

> The entire significance level (α) is in one tail of the distribution.

# Example:

> H₀: μ = 100

> H₁: μ > 100 (right-tailed)

> OR

> H₁: μ < 100 (left-tailed)

Use When:
> You're only interested in deviation in one direction (e.g., "Is this drug more effective?" not "Is it different?").

# Two-Tailed Test:

> Tests for an effect in either direction.

> The alternative hypothesis (H₁) states that a parameter is not equal
to a certain value.

> The significance level (α) is split between both tails of the distribution.

# Example:

> H₀: μ = 100

> H₁: μ ≠ 100

Use When:
> You care about any significant difference, regardless of direction (e.g., "Does this drug have a different effect, better or worse?").


| Feature                | One-Tailed Test         | Two-Tailed Test           |
| ---------------------- | ----------------------- | ------------------------- |
| Direction of test      | One direction           | Both directions           |
| Alternative hypothesis | μ > μ₀ or μ < μ₀        | μ ≠ μ₀                    |
| Rejection region       | One end of distribution | Both ends of distribution |
| α distribution         | All α in one tail       | α/2 in each tail          |

# Q 8: What is the Z-test, and when is it used in hypothesis testing?

Where:

> 𝜎 = population standard deviation

> 𝑛 = sample size

# Q 9: How do you calculate the Z-score, and what does it represent in hypothesis testing?

ANS: A Z-score (or standard score) tells you how many standard deviations a data point (or sample statistic) is from the population mean under the null hypothesis.

In hypothesis testing, it helps you determine how extreme your sample result is compared to the expected distribution, assuming the null hypothesis is true.

# Z-score Formula:

For sample mean (one-sample Z-test):

>> $Z = \frac{n}{\sigma} (\bar{X} - \mu)$

Where:

> 𝑋ˉ = sample mean

> 𝜇 = population mean (under 𝐻0)

> 𝜎 = population standard deviation

> 𝑛 = sample size

# What Does the Z-score Represent?

> A positive Z-score: Sample mean is above the population mean.

> A negative Z-score: Sample mean is below the population mean.

> A Z-score near 0: Sample mean is close to the population mean.

> A large absolute Z-score (e.g., |Z| > 1.96 for 95% confidence): Evidence against the null hypothesis.

# Example:

You want to test if a population mean is 100.

> You take a sample of size 36, get a sample mean of 103, and the known σ = 6.

>> $$
Z = \frac{\sqrt{36}}{6} \cdot \frac{103 - 100}{1} = \frac{1}{3} = 3
$$

> Z = 3 → The sample mean is 3 standard deviations above the hypothesized mean.

> This is statistically significant at the 0.01 level (p < 0.01), so you may reject 𝐻0.

# Interpretation in Hypothesis Testing:

> Compute Z-score using your sample data.

> Compare it to the critical Z-value (e.g., ±1.96 for 95% confidence).

> If Z falls in the rejection region, reject the null hypothesis.

> Alternatively, use the P-value corresponding to the Z-score.

# Q 12: What is the T-test, and how is it used in hypothesis testing?

ANS: The T-test is a statistical hypothesis test used to determine whether there is a significant difference between:

> A sample mean and a known population mean, or

> Two sample means, or

> Two related measurements (before and after, paired samples).

#It is especially useful when:

> The sample size is small (typically n < 30),

> The population standard deviation (σ) is unknown,

> The data is approximately normally distributed.

# Types of T-tests:

| **Type**                            | **Purpose**                                                                         |
| ----------------------------------- | ----------------------------------------------------------------------------------- |
| **One-sample T-test**               | Compares the sample mean to a known or hypothesized population mean.                |
| **Independent T-test** (Two-sample) | Compares means of **two independent groups**.                                       |
| **Paired T-test**                   | Compares means from **the same group at different times** (e.g., before vs. after). |

# How is the T-test used in hypothesis testing?

1. State the hypotheses:

> Null hypothesis (H₀): No difference exists (e.g., μ = μ₀)

> Alternative hypothesis (H₁ or Hₐ): A difference exists (e.g., μ ≠ μ₀)

2. Choose significance level (α):

> Common values: 0.05, 0.01

3. Calculate the test statistic:

>> $$
t = \frac{\bar{X} - \mu_0}{\frac{s}{\sqrt{n}}}
$$

> Xˉ : sample mean

> 𝜇0 : hypothesized population mean

> 𝑠 : sample standard deviation

> 𝑛 : sample size

4. Find the degrees of freedom (df):

> 𝑑𝑓 = 𝑛−1 (for one-sample)

5. Determine the critical value or P-value from the T-distribution.

6. Make a decision:

> If $$
|t| > t_{\text{critical}} \Rightarrow \text{Reject } H_0
$$

> Or, if P-value < α → reject H₀

# Example:

You want to test if the average sleep time of students is different from 8 hours.

> Sample mean (𝑋ˉ) = 7.6 hrs

> Hypothesized mean (𝜇0) = 8 hrs

> Sample std dev (s) = 0.5

> n = 16

>> $$
t = \frac{\frac{0.5}{\sqrt{16}}}{7.6 - 8} = \frac{0.125}{-0.4} = -3.2
$$

Compare this with the critical t-value for df = 15 at α = 0.05.

> If $$
|t| > t_{\text{critical}} \Rightarrow \text{Reject } H_0
$$


# Q 13: What is the relationship between Z-test and T-test in hypothesis testing?

ANS: The Z-test and T-test are closely related statistical methods used to test hypotheses about population means, but they apply in different situations based on data characteristics.

# Core Relationship:

| **Feature**                | **Z-test**                                                    | **T-test**                                       |
| -------------------------- | ------------------------------------------------------------- | ------------------------------------------------ |
| **Purpose**                | Tests hypotheses about population means                       | Same purpose                                     |
| **Formula Structure**      | Similar (difference of means / standard error)                | Same structure                                   |

$$

> Set up null (H₀) and alternative (H₁) hypotheses

> Choose significance level (α)

> Compute test statistic (Z or t)

> Compare with critical value or P-value

> Make a decision (reject or fail to reject H₀)

> The T-distribution approaches the Z-distribution as the sample size increases.

So with large samples, both tests give similar results.

| Situation             | Use                                 |
| --------------------- | ----------------------------------- |

# Q 14: What is a confidence interval, and how is it used to interpret statistical results?

ANS: A confidence interval (CI) is a range of values that is likely to contain the true population parameter (like the mean or proportion) with a certain level of confidence.

It gives an estimated range rather than a single value (point estimate), providing more information and helping assess the uncertainty of your estimate.

# Formula for Confidence Interval (for Mean):

> When population standard deviation σ is known (use Z):

>> $$
\text{CI} = \bar{X} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
$$

> When σ is unknown (use T):

>> $$
\text{CI} = \bar{X} \pm t_{\alpha/2} \cdot \frac{s}{\sqrt{n}}
$$

Where:

> Xˉ = sample mean

> σ or 𝑠 = population or sample standard deviation

> 𝑛 = sample size

> $$
Z_{\alpha/2}, \quad t_{\alpha/2}
$$ = Z or T value for the desired confidence level (e.g., 1.96 for 95%)

# How Confidence Intervals Are Used:

1. Estimate Population Parameters:

> CI provides a range where we believe the true mean/proportion lies.

2. Support Hypothesis Testing:

> If a confidence interval does not contain the null hypothesis value, we may reject the null.

>> Example: If H₀: μ = 100 and the 95% CI is [102, 108], we reject H₀.

3. Express Uncertainty:

> Narrower CI → more precise estimate

> Wider CI → more uncertain due to small sample or large variability

# Example:

A sample of 25 students has:

> Xˉ =70,

> 𝑠 = 10,

> 95% confidence level

>> $$
\text{CI} = 70 \pm t_{0.025,\,24} \cdot \frac{10}{\sqrt{25}} = 70 \pm 2.064 \cdot 2 = 70 \pm 4.128
$$


>> $$
\text{CI} = [65.87,\ 74.13]
$$

> Interpretation: We are 95% confident the true average score lies between 65.87 and 74.13.

# Q 16: How is Bayes' Theorem used in statistics, and what is its significance?

ANS:
# What is Bayes’ Theorem?
Bayes' Theorem is a fundamental concept in probability and statistics that describes how to update the probability of a hypothesis as new evidence or information becomes available.

# Formula:

>> $$
P(A \mid B) = \frac{P(B \mid A) \cdot P(A)}{P(B)}
$$

Where:

> P(A∣B): Posterior probability — probability of event A given B is true

> P(B∣A): Likelihood — probability of observing B if A is true

> P(A): Prior probability — initial belief about A before seeing B

> P(B): Marginal probability — total probability of observing B

# How is it used in statistics?

1. Statistical Inference:

> Updates estimates of population parameters using sample data.

> Core of Bayesian statistics, which treats parameters as random variables with probabilities.

2. Medical Testing:

> Estimate the probability of having a disease given a positive test result.

3. Spam Detection:

> Calculate the probability that an email is spam based on its content.

4. Decision Making Under Uncertainty:

> Helps in revising probabilities when new evidence becomes available (e.g., diagnostics, finance, machine learning).

#  Example (Medical Testing):

> 1% of people have a disease: 𝑃(𝐷) = 0.01

> Test is 99% accurate:

>> P(Positive∣D)=0.99

>> P(Positive∣No D)=0.01

You test positive. What is the probability you actually have the disease?

>> $$
P(D \mid \text{Positive}) = \frac{P(\text{Positive} \mid D) \cdot P(D)}{P(\text{Positive})}
$$

Where:

> P(Positive)=P(Positive∣D)⋅P(D)+P(Positive∣No D)⋅P(No D)

> =(0.99)(0.01)+(0.01)(0.99)=0.0099+0.0099=0.0198

> P(D∣Positive)= 0.0099/0.0198 = 0.5

Result: Even with a 99% accurate test, you only have a 50% chance of having the disease if you test positive — because the disease is rare.

# Significance of Bayes' Theorem:

> Provides a formal mechanism for learning from data.

> Bridges prior knowledge and new evidence.

> Foundation of Bayesian statistics, used in:

>> Machine Learning

>> Predictive Modeling

>> Artificial Intelligence

>> Real-time decision systems

>> Compares the means across groups defined by two independent variables (factors), and also tests for interaction effects.

> Example:

>> Studying the effect of exercise type and diet on weight loss.

Used when you're examining the combined effect of two factors and their interaction on a dependent variable.

3. Repeated Measures ANOVA

> Purpose:

>> Compares means when the same subjects are measured multiple times (within-subjects design).

> Example:

>> Measuring blood pressure of patients before, during, and after treatment.

Used for time-series or longitudinal studies involving repeated observations.

4. Mixed-Design ANOVA (Split-Plot ANOVA)

> Purpose:

>> Combines both within-subjects (repeated measures) and between-subjects (independent groups) factors.

> Example:

>> Comparing two drug treatments over three time points on the same participants.

Used in complex experimental designs with both repeated and independent factors.

5. MANOVA (Multivariate Analysis of Variance)

> Purpose:

>> Similar to ANOVA but compares multiple dependent variables at once.

> Example:

>> Studying how teaching methods affect both math and reading scores.

Used when there are multiple continuous outcomes and you want to consider their correlation.

6. ANCOVA (Analysis of Covariance)

> Purpose:

>> ANOVA that includes one or more covariates (continuous variables that might influence the dependent variable).

> Example:

>> Examining the effect of a training program on test scores while controlling for prior knowledge.

Adjusts for variables that could confound the results.

# Q 22: What is the F-test, and how does it relate to hypothesis testing?

ANS:
# What is the F-test?
The F-test is a statistical test used to compare two variances or to evaluate whether group means differ significantly by comparing variability between groups to variability within groups.

It is based on the F-distribution, named after Sir Ronald Fisher.

# How the F-Test Works?

The F-test calculates an F-statistic, which is a ratio of two variances:

>> $F = \frac{\text{Variance between groups}}{\text{Variance within groups}}$

> If the group means are similar, the between-group variance is small, and F ≈ 1.

> A larger F value suggests significant differences among group means.

# F-test in Hypothesis Testing:
The F-test is used primarily in:

1. ANOVA (Analysis of Variance)

> Tests if three or more group means are equal.

> Hypotheses:

>> Null (H₀): All group means are equal.

>> Alternative (H₁): At least one group mean is different.

> If the calculated F is greater than the critical F value, or if p-value < α, we reject H₀.

2. Comparing Two Variances

> Hypotheses:

>> H₀: Variance₁ = Variance₂

>> H₁: Variance₁ ≠ Variance₂ (or > or <)

> Used in quality control or before a t-test (to check equal variances).

# Formula for F in One-Way ANOVA:

>> $$
F = \frac{MS_{\text{between}}}{MS_{\text{within}}}
$$

Where:

> $MS_{\text{between}}$ = Mean square between groups

> $MS_{\text{within}}$ = Mean square within groups (error)

| ----------------- | ------------------------------------------------- |
| **Purpose**       | Compare variances or group means                  |
| **Distribution**  | F-distribution (asymmetric, non-negative)         |
| **Used In**       | ANOVA, testing variance equality                  |
| **Key Statistic** | $F = \frac{\text{Variance}_1}{\text{Variance}_2}$ |
| **Decision Rule** | Reject H₀ if $F$ is large or $p < \alpha$         |
